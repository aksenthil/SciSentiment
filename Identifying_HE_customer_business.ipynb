```python
def getUniqueItems(iterable):
    seen = set()
    result = []
    for item in iterable:
        if item not in seen:
            seen.add(item)
            result.append(item)
    return result
 
```

```python
dup=R_data['Applicant business'].tolist()
results = []
for ak in dup:
    hj=getUniqueItems(ak)
    results.append(hj)
R_data['Business_of_applicant']=  results 
```

```python
# function for calculating the frequency 
def freq(str): 
  
    # break the string into list of words 
    str_list = str.split() 
  
    # gives set of unique words 
    unique_words = set(str_list) 
      
    for words in unique_words : 
        print(words , ':',str_list.count(words)) 
  
```
